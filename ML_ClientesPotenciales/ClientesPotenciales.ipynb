```python
import pandas as pd
import numpy as np
```

```python
#Importo los datos
df = pd.read_csv('Leads.csv')
df.head().T
```

```python
df.dtypes
```

```python
#Voy a verificar los valores de las columnas con .value_counts() para identificar qu columnas son innecesarias
col = list(df.columns)
for i in col:
    print(f"----------------------------------- {i} ------------------------------------------")
    print(df[i].value_counts(dropna=False),"\n\n") 
```

```python
#Elimino las columnas que considero inservibles, porque tienen muchos Nan, o porque solo tienen un valor mayoritario o Select,
#que corresponde a cuando no seleccionas nada en un formulario.

df.drop(['Prospect ID','Lead Number', 'Do Not Call', 'Country', 'How did you hear about X Education',
          'What matters most to you in choosing a course', 'What matters most to you in choosing a course',
           'Search', 'Magazine', 'Newspaper Article', 'X Education Forums', 'Newspaper', 'Digital Advertisement',
           'Through Recommendations', 'Receive More Updates About Our Courses', 'Lead Quality',
           'Update me on Supply Chain Content', 'Get updates on DM Content', 'Lead Profile', 'Asymmetrique Activity Index',
           'Asymmetrique Profile Index', 'Asymmetrique Activity Score', 'Asymmetrique Profile Score',
           'I agree to pay the amount through cheque'], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
#Voy a convertir las variables si o no en 1 y 0

varlist =  ['Do Not Email', 'A free copy of Mastering The Interview']

def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

df[varlist] = df[varlist].apply(binary_map)
```

```python
df.head(5)
```

```python
#El resto de valores Select los convierto en Nan, para poder eliminarlos posteriormente.

df = df.replace(["Select"], np.nan)
```

```python
#Reviso los valores nulos.

df.isnull().sum().sort_values(ascending=False)
```

```python
#Vuelvo a repasar las columnas con ms datos nulos

c = ["Tags", "What is your current occupation", "Specialization", "City"]
col = list(df[c])
for i in col:
    print(f"----------------------------------- {i} ------------------------------------------")
    print(df[i].value_counts(dropna=False),"\n\n")
```

```python
#Los NaN de 'City' los voy a convertir a Mumbai, ya que posiblemente sean todos de all.

df["City"] = df["City"].replace(np.nan, "Mumbai")

print(df["City"].value_counts(dropna=False),"\n\n")
```

```python
#Lo mismo con los NaN de 'What is your current occupation', seguramente pertenezcan todos a Unemployed, los convierto.

df["What is your current occupation"] = df["What is your current occupation"].replace(np.nan, "Unemployed")

print(df["What is your current occupation"].value_counts(dropna=False),"\n\n")
```

```python
#El resto de valores nulos los elimino del Dataframe

df = df.dropna()

df.isnull().sum().sort_values(ascending=False)
```

```python
#Identifico las columnas categricas y las columnas numricas

from sklearn.compose import make_column_selector as selector

numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)

numerical_columns = numerical_columns_selector(df)
categorical_columns = categorical_columns_selector(df)
```

```python
df[categorical_columns].nunique()
```

```python
df[numerical_columns].nunique()
```

```python
#Convierto el dataframe a minsculas y quito los espacios, adems, las variables object las convierto en string.

replacer = lambda str: str.lower().str.replace(' ', '_')
df.columns = replacer(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)
y_train = df_train.converted.values
y_val = df_val.converted.values

del df_train['converted']
del df_val['converted']

df_train.head().T
```

# Ingeniera de propiedades

```python
train_dict = df_train.to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

# Entrenamiento del modelo

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
val_dict = df_val.to_dict(orient='records')
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
converted = y_pred >= 0.5
converted
```

```python
round((y_val == converted).mean(), 3)
```

# Serializacin del modelo

```python
import pickle
```

```python
with open('modelo/converted-modelo.pck', 'wb') as f:
    pickle.dump((dv, model), f)
```

```python
with open('modelo/converted-modelo.pck', 'rb') as f:
    dv, model = pickle.load(f)
    X_val = dv.transform(val_dict)
    y_pred = model.predict_proba(X_val)

y_pred
```
